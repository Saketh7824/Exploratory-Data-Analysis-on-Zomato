```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('darkgrid')
import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv('C:/Data Sets/zomato.csv')
```

```python
data.head()
```

```python
data.shape
```

```python
data.columns.values
```

```python
data=data.drop(['url','address','phone','menu_item','dish_liked','reviews_list'],axis=1)
```

```python
data.head()
```

```python
data.info()
```

```python
data.drop_duplicates(inplace=True)
```

```python
data.shape
```

```python
data['rate'].unique()
```

```python
def handlerate(value):
    if(value=='NEW' or value=='-'):
        return np.nan
    else:
        value=str(value).split('/')
        value=value[0]
        return float(value)
```

```python
data['rate']=data['rate'].apply(handlerate)
```

```python
data['rate'].head()
```

```python
data['rate'].mean()
```

```python
#filling null values in rate columns with mean
data['rate'].fillna(data['rate'].mean(),inplace=True)
data['rate'].isnull().sum()
```

```python
data.info()
```

```python
# Droping Null Values
data.dropna(inplace=True)
```

```python
data.head()
```

```python
data.rename({'approx_cost(for two people)':'cost2plates','listed_in(type)':'type'},axis=1,inplace=True)
```

```python
data.head()
```

```python
data['location'].unique()
```

```python
data['listed_in(city)'].unique()
```

```python
data=data.drop(['listed_in(city)'],axis=1)
```

```python
data['cost2plates'].unique()
```

```python
def remove_comma(val):
    val=str(val)
    if ',' in val:
        val=val.replace(',','')
        return float(val)
    else:
        return float(val)
```

```python
data['cost2plates']=data['cost2plates'].apply(remove_comma)
```

```python
data['cost2plates'].unique()
```

```python
data.head()
```

```python
#Cleaning rest type columns
rest_types=data['rest_type'].value_counts(ascending=True)
rest_types
```

```python
rest_types_lessthan1000=rest_types[rest_types<1000]
rest_types_lessthan1000
```

```python
#Making rest types less than 1000 in frequency as others
def handling_rest_types(val):
    if val in rest_types_lessthan1000:
        return 'others'
    else:
        return val
```

```python
data['rest_type']=data['rest_type'].apply(handling_rest_types)
```

```python
data['rest_type'].value_counts()
```

```python
data.head()
```

```python
#cleaning location column
location=data['location'].value_counts()
location
```

```python
location_lessthan300=location[location<300]
location_lessthan300
```

```python
def handle_location(val):
    if val in location_lessthan300:
        return 'others'
    else:
        return val
```

```python
data['location']=data['location'].apply(handle_location)
data['location'].value_counts()
```

```python
#cleaning cuisines columns
cuisine=data['cuisines'].value_counts(ascending=False)
cuisine
```

```python
cuisine_lessthan100=cuisine[cuisine<100]
cuisine_lessthan100
```

```python
def cuisines_handling(val):
    if val in cuisine_lessthan100:
        return 'others'
    else:
        return val
```

```python
data['cuisines']=data['cuisines'].apply(cuisines_handling)
data['cuisines'].value_counts()
```

```python
data.head()
```

```python
data['type'].value_counts()
```

## Visualizations

```python
#countplot of various locations
plt.figure(figsize=(16,10))
sns.countplot(data['location'],linewidth=1,edgecolor='black',palette='muted')
plt.xticks(rotation=90)
plt.show()
```

```python
#visualizing the online order
plt.figure(figsize=(8,8))
sns.countplot(data['online_order'],linewidth=1,edgecolor='black',palette='Set2')
plt.show()
```

```python
#visualizing the book table
plt.figure(figsize=(6,6))
sns.countplot(data['book_table'],palette='muted',linewidth=1,edgecolor='black')
plt.show()
```

```python
#visualizing the online order vs rate
plt.figure(figsize=(6,6))
sns.boxplot(data['online_order'],data['rate'],palette='deep')
plt.show()
```

```python
#visualizing the book table vs rate
plt.figure(figsize=(6,6))
sns.boxplot(x='book_table',y='rate',data=data,palette='colorblind')
plt.show()
```

```python
# visualize online order facility,location wise
df1=data.groupby(['location','online_order'])['name'].count()
df1.to_csv('location_online.csv')
df1=pd.read_csv('location_online.csv')
df1=pd.pivot_table(df1,values=None,index=['location'],columns=['online_order'],fill_value=0,aggfunc=np.sum)
df1
```

```python
df1.plot(kind='bar',figsize=(15,8),edgecolor='black')
```

```python
## visualize book table facility,location wise
df2=data.groupby(['location','book_table'])['name'].count()
df2.to_csv('location_booktable.csv')
df2=pd.read_csv('location_booktable.csv')
df2=pd.pivot_table(df2,values=None,index=['location'],columns=['book_table'],fill_value=0,aggfunc=np.sum)
df2
```
